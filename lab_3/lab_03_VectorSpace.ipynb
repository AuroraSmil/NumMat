$$

The operation of addition and multiplication in the space of real polynomials is defined coefficient-wise. For $p(x) = a_0 + a_1 x + ... + a_n x^n$ and $q(x) = b_0 + b_1 x + ... + b_n x^n$, $a_0, a_1, ... , a_n, b_0, b_1, ... , b_n \in \mathbb{R}$ , we define 
$$ (p + q) (x) = (a_0 + b_0) + (a_1 + b_1) x + ... + (a_n + b_n) x^n $$ and $$ (\lambda p) (x) = \lambda a_0 + \lambda a_1 x + ... + \lambda a_n x^n, \lambda \in \mathbb{R}$$

### d)
Write down the definition of a norm on a vector space.

<font color ='blue'>
    Solution:    
</font>

A *norm* on a vector space $X$ is a function $\|\,.\| : X \to \mathbb{R}$, such that for all $x,y\in\mathbb{R}$:

1. *Positivity:* $0 \leq \|x\| \leq\infty$ and $\|x\| = 0$ if and only if $x=0$.
2. *Homogenity:* $\|\alpha x\| = |\alpha|\|x\|$ for any $\alpha\in\mathbb{C}$.
3. *Triangle inequality:* $\|x + y\| \leq \|x\| + \|y\|$

### e)
Let $x_0,...,x_n$ be $n+1$ different real numbers. For a polynomial $q \in \mathbb{P}_n(\mathbb{R})$ we now define
$$
\|q\|_P := \sum_{k=0}^n|q(x_k)|.
$$
Show that $\|\cdot\|_P$ is a norm on $\mathbb{P}_n(\mathbb{R}).$



<font color ='blue'>
    Solution:
    
</font>

Let $p, q \in\mathbb{P}_n(\mathbb{R})$.

*Positivity*

Since the norm defined above is a sum of absolute values, the norm is always a positive number. If the polynomial $q(x)$ is the null element, so $q(x)=0$, we se that the norm is also zero. From the other side, if we have $\|q\| =0$ for a arbitrary polynomial $q$, we must have $q(x_i)=0$ for $i=0, 1, ...,n$. From the fundamental theorem of algebra we know that a polynomial of degree $n$ can have at most $n$ roots, unless it is the zero. Therefore we must have that $\|q\|=0$ implies $q=0$.

*Homogenity*
$$
\begin{align}
\|\alpha q\|_P &= \sum_{k=0}^n|\alpha q(x_k)| = \sum_{k=0}^n|\alpha|| q(x_k)|= |\alpha|\sum_{k=0}^n| q(x_k)| = |\alpha|\|q\|
\end{align}
$$

*Triangle inequality*
$$
\begin{align}
\|p +q\|_P &= \sum_{k=0}^n|(p  + q)(x_k)| = \sum_{k=0}^n|p(x_k)+ q(x_k)|\leq \sum_{k=0}^n(| p(x_k)|+| q(x_k)|) = \sum_{k=0}^n| p(x_k)| + \sum_{k=0}^n| q(x_k)| = \|p\| + \|q\|
\end{align}
$$

Thus the norm axioms holds for any $p,q$, so the norm $\|\,.\|_P$ is a norm on $\mathbb{P}_n(\mathbb{R})$.
